```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax=plt.axes()
sns.heatmap(books.isna(),cbar=False,ax=ax)
plt.xlabel("Column in the dataframe")
plt.ylabel("Row entry in the dataframe")
plt.show()
```

- Look at the columns along the bottom (title, subtitle, authors etc.)
- And now, vertically up from each column
- Each black horizontal line in a column shows that a book entry (row) is missing that value

```python
import numpy as np
# create a column called missing_description, where each row has value 1 if it doesn't have a description (true), else 0 (false)
books["missing_description"]=np.where(books["description"].isna(),1,0)

# convert published_year column to the age_of_book
books["age_of_book"]=2025-books["published_year"]
```

```python
# select the columns of interest
columns_of_interest=["num_pages","age_of_book","missing_description","average_rating"]

# create a correlation matrix with columns_of_interest using Spearman's method
# Pearson's method is more suitable for continuous var, but here we have non-continuous values
correlation_matrix=books[columns_of_interest].corr(method="spearman")

# convert the matrix into a seaborn heatmap
sns.set_theme(style="white")
# plt.figure(figsize(8,6))
heatmap=sns.heatmap(correlation_matrix,annot=True,fmt=".2f",cmap="coolwarm",cbar_kws={"label":"Spearman Correlation"})
heatmap.set_title("Correlation heatmap")
plt.show()
```

- We have to observe the correlation of missing_description with other variables, to find out if any variable has high correlation (r) with missing_description
- (eg) if num_pages has high r, it means that either very long or very short books have missing description
- We can see from the plot that all columns have somewhat decent r values with missing_desc, no extreme values
- Maybe we can delete the books with missing_description

```python
# Show all books with missing num_pages or published_year or description or average rating
books[(books["num_pages"].isna()) | (books["published_year"].isna()) | (books["description"].isna()) | (books["average_rating"].isna())]
```

- There are only 303 such books, which is quite small compared to the total rows (6810)
- So we can neglect these books\rows


```python
# new dataframe book_missing that contains only the rows that have all the values
book_missing=books[~(books["num_pages"].isna()) & ~(books["published_year"].isna())  & ~(books["description"].isna()) & ~(books["average_rating"].isna())]
book_missing # contains 6810 - 303 = 6507 rows
```

```python
# Now let's consider categories column
# From column stats, categories has about 530 distinct values
# Let's see how categories is distributed
book_missing["categories"].value_counts().reset_index().sort_values("count",ascending=False)
```

- Clearly, the distribution is uneven (chart view of above table (x-axis: categories, y-axis: count))
- Let's deal with it later using LLMs (I think)

```python
# Now take a look at the descriptions column again
# The description should be meaningful for the book, else it's useless
# In book_missing, some descriptions are pretty useless, like "Donation."
# We should get rid of them
# First create a column called words_in_description that stores the length of it
book_missing["words_in_description"]=book_missing["description"].str.split().str.len()
book_missing
```

```python
# Bin 1: Show the description of the books whose missing_description is between 1 and 4
book_missing.loc[book_missing["words_in_description"].between(1,4),"description"]
```

- Clearly, these descriptions are useless

```python
# Bin 2: Show the description of the books whose missing_description is between 5 and 14
book_missing.loc[book_missing["words_in_description"].between(5,14),"description"]
```

- Not very helpful either

```python
# Bin 3: Show the description of the books whose missing_description is between 15 and 24
book_missing.loc[book_missing["words_in_description"].between(15,24),"description"]
```

- Descriptions are getting better, but maybe not enough

```python
# Bin 4: Show the description of the books whose missing_description is between 25 nd 34
book_missing.loc[book_missing["words_in_description"].between(25,34),"description"]
```

- This is it!

```python
# Filter all rows with description >= 25 words
book_missing_25_words=book_missing[book_missing["words_in_description"]>=25]
book_missing_25_words
```

- Now we have about 5200 rows (reduced from 6507)
- Now, look at the subtitles column (some have it, some don't)
- We can create a column called title_and_subtitle to combine them if they both exist, else just keep the title

```python
# If both title and subtitle exist, join them, else just keep the title in the new column
book_missing_25_words["title_and_subtitle"]=(
    np.where(book_missing_25_words["subtitle"].isna(),book_missing_25_words["title"],
                book_missing_25_words[["title","subtitle"]].astype(str).agg(": ".join,axis=1))
)
book_missing_25_words
```

```python
# Now let's create a new column tagged_description that combines the isbn and description column (why...? Maybe I'll know later)
book_missing_25_words["tagged_description"]=book_missing_25_words[["isbn13","description"]].astype(str).agg(" ".join,axis=1)
book_missing_25_words
```

```python
# Now, we can drop unneeded columns and save
(book_missing_25_words
    .drop(["subtitle","missing_description","age_of_book","words_in_description"],axis=1)
    .to_csv("books_cleaned.csv",index=False)
 )
```
